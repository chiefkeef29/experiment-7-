```python
#reverse of a number
num = int(input("enter a chain of numbers"))
rev = 0
while num > 0:
  digit = num % 10
  rev = rev*10 + digit
  num //=10
print("reversed number is:", rev)
```

```python
#check palindrome
st = input("enter the string:")
rev = st[::-1]
if st == rev:
  print("palindrome")
else:
  print("not palindrome")
```

```python
s = "anurag"
i,j = 0, len(s) - 1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
   is_palindrome = False
  break
  i += 1
  j -= 1
if is_palindrome:
  print("palindrome")
else:
  print("not a palindrome")
```

```python
#a program to choose num from list
num = [10,20,30,40,50]
key = int(input("enter element to search"))
i = 0
while i < len(num):
  if num[i] == key:
    print("element found at index" , i)
    break
  i += 1
else:
    print("element not found")

```
